```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import spherical_jn
from scipy.integrate import quad
from scipy.interpolate import interp1d
```

$k_\mathrm{min} = k_\mathrm{f} = 2\pi  / L_\mathrm{box}$  
$k_\mathrm{max} = k_\mathrm{Nyq} = \pi N_\mathrm{mesh} / L_\mathrm{box}$

```python
Lbox, Nmesh = 3500., 1024
print (2*np.pi / Lbox, np.pi * Nmesh / Lbox)
```

```python
colors = ['C%s' % i for i in range(3)]

fig1, ax1 = plt.subplots(2,2, figsize=(12, 4))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

fig2, ax2 = plt.subplots(2,2, figsize=(12, 4))
plt.subplots_adjust(wspace=0.2, hspace=0.4)

fig3, ax3 = plt.subplots(2,2, figsize=(12, 4), sharey=True)
plt.subplots_adjust(wspace=0.2, hspace=0.4)

for i, red in enumerate(['lowz', 'cmass']):
    if 'lowz' in red: zbin = '0p2-0p43'
    elif 'cmass' in red: zbin = '0p43-0p7'
    for j, sky in enumerate(['sgc', 'ngc']):
        
        # Nbodykit on randoms
        if 'lowz' in red and 'sgc' in sky: norm, alpha = 95.29458797698169, 0.020396601298
        elif 'lowz' in red and 'ngc' in sky: norm, alpha = 243.8557397248968, 0.020160886971
        elif 'cmass' in red and 'sgc' in sky: norm, alpha = 204.77085121873222, 0.0198743279313
        elif 'cmass' in red and 'ngc' in sky: norm, alpha = 571.4117399094032, 0.0191803524419
        
        idx_kcut, kmins = [], []
        for n, (Lbox, sf) in enumerate(zip([100000, 30000, 3500, 1000], [2., 2., 8., 12.])):
            Nmesh = 1024
            windata = np.loadtxt(os.path.join('Qs_Lbox%s', 'window_boss_%s_%s.dat') % (Lbox, red, sky))
            k_i = windata[:, 0]
            win_i =  windata[:, 1:].T
            if n == 0:
                kmin, kmax_0 = 2*np.pi / Lbox, np.pi * Nmesh / Lbox
                kmax_0 /= sf # security factor
                kmask = np.where((k_i < kmax_0))[0]
                k, win = k_i[kmask], win_i[:, kmask]
            else:
                kmin, kmax = 2*np.pi / Lbox, np.pi * Nmesh / Lbox
                kmax /= sf
                kmask = np.where((kmax_0 < k_i) & (k_i < kmax))[0]
                kmax_0 = kmax
                k, win = np.concatenate((k, k_i[kmask]), axis=-1), np.concatenate((win, win_i[:, kmask]), axis=-1)
            idx_kcut.append(len(k)-1)
            kmins.append(kmin)
        kmins = np.array(kmins)
        
        ax1[i,j].set_title('%s %s' % (red, sky))
        for l, (w, c) in enumerate(zip(win,colors)):
            ax1[i,j].plot(k, w, '%s' % c)
        ax1[i,j].grid()
        ax1[i,j].set_xscale('log')
        ax1[i,j].set_xlabel(r'$k$')

        ax2[i,j].set_title('%s %s' % (red, sky))
        for l, (w, c) in enumerate(zip(win,colors)):
            if l == 0:
                ax2[i,j].plot(k, k**2 * w, '%s.' % c)
                ax2[i,j].plot(k[idx_kcut], (k**2 * w)[idx_kcut], 'r.')
                ax2[i,j].plot(kmins, np.zeros_like(kmins), 'gx')
        ax2[i,j].grid()
        ax2[i,j].set_xscale('log')
        ax2[i,j].set_xlabel(r'$k$')
        # ax2[i,j].set_xlim(1e-4, 1e-1)
        #ax2[i,j].set_ylim(0, 7e6)
        
        # Configuration-space window
        s = np.logspace(-2, 6, 2048)
        klog = np.geomspace(k[1], k[-1], 1024)
        kk, ss = np.meshgrid(klog, s, indexing='ij')
        ww = interp1d(k, win, axis=-1)(kk)
        jn = np.array([(-1)**(l//2) * spherical_jn(l, kk*ss) for l in [0, 2, 4]])
        qq = np.trapz(np.einsum('ks,wks,wks->wks', kk**2, ww, jn), x=klog, axis=1) / (2*np.pi**2)
        
        normalpha = norm / alpha
        
        ax3[i,j].set_title('%s %s' % (red, sky))
        for l, (q, c) in enumerate(zip(qq,colors)):
            ax3[i,j].plot(s, q / normalpha, '%s' % c)
            ax3[i,j].grid()
        ax3[i,j].set_xscale('log')
        ax3[i,j].set_ylim(-0.5, 1)
        ax3[i,j].set_xlim(1e1, 1e4)
        ax3[i,j].set_xlabel(r'$s$')
        
        # normalization
        ii = interp1d(k, k**2 * win[0] * spherical_jn(0, 0), kind='cubic')
        q0quad = quad(ii, k[1], k[-1])[0] / (2*np.pi**2)
        
        print ('norm %s %s: %.3f, (quad: %.3f)' % (sky, red, qq[0,0] / normalpha, q0quad / normalpha))
        
        filename = 'window_boss_%s_%s_ell024_2022.dat' % (red, sky)
        np.savetxt(os.path.join('Qw', filename), np.vstack([s, np.vstack([q / normalpha for q in qq])]).T,
                 header='s [Mpc/h], q0, q2, q4', fmt='%.6e')
        
```
